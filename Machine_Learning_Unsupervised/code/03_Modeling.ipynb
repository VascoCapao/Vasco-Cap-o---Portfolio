# Final Model
After analyzing and testing several models, what we were able to identify as having the most potential was the Hierarchical model with Robust Scaler and also using the variables created from percentages

```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.cluster import  AgglomerativeClustering

import os

# Suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing the dataset with all the customer information
customer_info_final  = pd.read_csv('customer_info_precleaned.csv')
```

```python
customer_info_final 
```

```python
# Remove columns for clustering
columns_to_remove = ['customer_id','customer_name', 'customer_gender', 'loyalty_card_number', 'latitude', 'longitude', 'degree']
customer_for_clustering = customer_info_final.drop(columns=columns_to_remove)
```

```python
customer_for_clustering
```

```python
#### Scaling the data
customer_info_scaled = RobustScaler().fit_transform(customer_for_clustering)
```

```python
# Creating a new variable with cluster labels using AgglomerativeClustering
customer_for_clustering['cluster_ward'] = AgglomerativeClustering(
    linkage='ward', n_clusters=7
    ).fit_predict(customer_info_scaled)
```

```python
# Set the display option to show all columns
pd.set_option('display.max_columns', None)

# Group the DataFrame by the 'cluster_ward' column and calculate the mean for each variable
customer_for_clustering.groupby(['cluster_ward']).mean()
```

# Customer Cluster Analysis

## Cluster 0: Diversifiers

## Cluster 1: Occasional discount shoppers

## Cluster 2: Large Families

## Cluster 3: Vegetarians

## Cluster 4: Pet Owners

## Cluster 5: Young Shoppers

## Cluster 6: Premium Clients


```python
# Creating a Dataframe with the clusters and the customer information
df_with_clusters = pd.concat([customer_info_final, customer_for_clustering['cluster_ward']], axis=1)
```

```python
# Define a dictionary mapping cluster IDs to their corresponding names
cluster_names = {
    0: "Diversifiers",
    1: "Occasional discount shoppers",
    2: "Large Families",
    3: "Vegetarians",
    4: "Pet Owners",
    5: "Young Shoppers",
    6: "Premium Clients"
}

# Map the cluster ward values to the labels
df_with_clusters['cluster_ward'] = df_with_clusters['cluster_ward'].map(cluster_names)
```

```python
# Save the DataFrame to a CSV file for further analysis
df_with_clusters.to_csv('df_with_clusters')
```

### Saving each cluster to a separate CSV file


```python
# Create a directory to save the CSV files
output_dir = "custom_cluster_csv_files"
os.makedirs(output_dir, exist_ok=True)

# Iterate over the unique clusters and export the corresponding data to CSV files
for cluster_id in df_with_clusters['cluster_ward'].unique():
    # Filter the data for the current cluster
    cluster_data = df_with_clusters[df_with_clusters['cluster_ward'] == cluster_id]
    # Get the name of the current cluster
    cluster_name = cluster_names.get(cluster_id, f"Cluster {cluster_id}")  # Default to "Cluster {cluster_id}" if name not found
    # CSV file name for the current cluster
    filename = f"{cluster_name}.csv"
    # Full path for the CSV file
    filepath = os.path.join(output_dir, filename)
    # Export the cluster data to the CSV file
    cluster_data.to_csv(filepath, index=False)

    print(f"{cluster_name} exported to {filename}")
```
